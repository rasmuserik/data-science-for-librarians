```python
import requests
import json
```

```python
title = "List_of_hip_hop_musicians"
result = json.loads(requests.get(
    'https://en.wikipedia.org/w/api.php', 
    params = {
        "action": "query",
        "titles": title,
        "prop": "links",
        "format": "json",
        "plnamespace": [0],
        "pllimit": 500
    }).text)
[link["title"] for link in list(result["query"]["pages"].values())[0]["links"]]
result
```

```python
list_of_artists = result["query"]["pages"]["140853"]["links"]
artists = [];
for artist in list_of_artists:
    print(artist["title"])
    artists.append(artist["title"])
```

```python
artists
```

```python
title= "Eminem"
artist_year = []
for title in artists:
    try:
        result = json.loads(requests.get(
            'http://dbpedia.org/sparql?default-graph-uri=http%3A%2F%2Fdbpedia.org&query=DESCRIBE%20%3Chttp%3A%2F%2Fdbpedia.org%2Fresource%2F' + title +'%3E&format=application%2Fjson-ld', 
            ).text)
        list(result.keys())
        year = result["http://dbpedia.org/resource/" + title]['http://dbpedia.org/ontology/activeYearsStartYear']
        print(title, year[0]["value"])
        artist_year.append([year[0]["value"], title])
    except:
        pass
```
